# Project 1 Section 2
import house prices dataset and filter warnings

```python
#precondition - data importing
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
pd.set_option('display.max_columns', None)
url = "https://raw.githubusercontent.com/mtsilimos/url/main/houseprices.csv"

dataset = pd.read_csv(url)
```

# Fulfill task 2
Remove the columns in your dataset which have less than 700 non-NaN values

```python
#task 2
dataset.dropna(axis=1, thresh = 700, inplace = True)
```

# Fulfill task 3
Perform median imputation on your dataset

```python
#task 3
dataset = dataset.fillna(dataset.median(numeric_only=True))
```

```python
dataset.info()
```

# Define a function to convert Y/N feature to numeric
feature CentralAir is converted from Y/N to 1/0

```python
#CentralAir convert yes and no
def binary_map(x):
    return x.map({"Y": 1, "N": 0})

# Applying the function to the dataset
dataset[['CentralAir']] = dataset[['CentralAir']].apply(binary_map)
```

# Fulfill task 5
Train a linear regression model on the dataset

```python
#task 5
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#task 5
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```
